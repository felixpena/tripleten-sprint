### Mira cuántos juegos fueron lanzados en diferentes años. ¿Son significativos los datos de cada período?

```python
# Contando la cantidad de juegos lanzados en cada año
games_per_year = df['year_of_release'].value_counts().sort_index()

# Visualizando los datos
games_per_year
```

```python
# Creando un gráfico de barras
plt.figure(figsize=(15, 8))
games_per_year.plot(kind='bar', color='skyblue')
plt.title('Cantidad de Juegos Lanzados por Año')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Juegos')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Mostrando el gráfico
plt.show()
```

Gráfico de barras que muestra la cantidad de juegos lanzados por año. Como se puede ver:

- La cantidad de lanzamientos aumenta significativamente a partir de principios de los años 2000, alcanzando su punto máximo alrededor de 2008-2009.
- Después de 2009, hay una tendencia a la baja en el número de lanzamientos anuales.

Este gráfico proporciona una visualización clara de cómo han evolucionado los lanzamientos de juegos a lo largo de los años. 

__Interesante, pero que impacto tuvo la industria de plataformas en la evolución__

```python
# Preparando los datos para el gráfico de líneas por plataforma

# Creamos un DataFrame pivotado para contar la cantidad de juegos por año y plataforma
pivot_df = df.pivot_table(index='year_of_release', columns='platform', values='name', aggfunc='count', fill_value=0)

# Creando el gráfico de líneas para cada plataforma
plt.figure(figsize=(20, 10))
for platform in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[platform], label=platform)

plt.title('Cantidad de Juegos Lanzados por Año y Plataforma')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Juegos')
plt.legend(loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)

# Mostrando el gráfico
plt.show()
```

Gráfico de líneas que muestra la cantidad de juegos lanzados por año y plataforma. En este gráfico:

Cada línea representa una plataforma diferente.
El eje horizontal muestra los años, y el eje vertical indica la cantidad de juegos lanzados.
La variación en la cantidad de lanzamientos por plataforma a lo largo del tiempo se visualiza claramente.

Este tipo de gráfico es útil para observar tendencias, como el surgimiento y la declinación de popularidad de diferentes plataformas a lo largo del tiempo, así como para comparar la actividad de lanzamiento entre plataformas en un año dado.

```python
# Creando un gráfico de área apilada para la cantidad de juegos lanzados por año y plataforma
plt.figure(figsize=(20, 10))
pivot_df.plot(kind='area', stacked=True, figsize=(20, 10))

plt.title('Cantidad de Juegos Lanzados por Año y Plataforma (Gráfico de Área Apilada)')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Juegos')
plt.legend(title='Plataforma', loc='upper left')
plt.xticks(rotation=45)

# Mostrando el gráfico
plt.show()
```

Gráfico de área apilada que muestra la cantidad de juegos lanzados por año y plataforma. En este gráfico:

- Cada área de color representa una plataforma diferente.
- El eje horizontal muestra los años, y el eje vertical indica la cantidad de juegos lanzados.
- Las áreas están apiladas, por lo que la altura total de todas las áreas en un punto dado representa el número total de juegos lanzados en ese año, mientras que la altura de cada área individual representa la contribución de esa plataforma específica.

Este tipo de gráfico es útil para visualizar no solo el volumen total de juegos lanzados cada año, sino también cómo la proporción de juegos lanzados en cada plataforma ha cambiado con el tiempo.

```python
# Identificando puntos representativos para poner las etiquetas
# Seleccionamos algunos puntos donde las plataformas tienen contribuciones significativas


# Creando el gráfico de nuevo
plt.figure(figsize=(20, 10))
pivot_df.plot(kind='area', stacked=True, figsize=(20, 10))

# Añadiendo etiquetas en puntos representativos
for year in pivot_df.index:
    y_offset = 0  # Desplazamiento inicial para la posición de las etiquetas
    for platform in pivot_df.columns:
        # Obtenemos la contribución de la plataforma para el año
        contribution = pivot_df.loc[year, platform]
        # Solo etiquetamos si la contribución es significativa
        if contribution > 50:
            plt.text(year, y_offset + contribution/2, platform, ha='center', va='center')
        y_offset += contribution

plt.title('Cantidad de Juegos Lanzados por Año y Plataforma (Gráfico de Área Apilada)')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Juegos')
plt.legend(title='Plataforma', loc='upper left')
plt.xticks(rotation=45)

# Mostrando el gráfico
plt.show()
```

Se agregaron algunas etiquetas representativas al gráfico de área apilada para mostrar la contribución de ciertas plataformas en años específicos. Las etiquetas se han colocado en puntos donde las plataformas tienen contribuciones significativas, y se ha intentado distribuirlas de manera que sean lo más legibles posible.

Debido a limitaciones de espacio y legibilidad, no todas las plataformas están etiquetadas en cada año. Además, el criterio para etiquetar fue una contribución mayor a 50 juegos en un año, pero esto es ajustable según lo que consideres significativo.

Este método puede ayudar a dar una idea de cuáles plataformas dominaron en ciertos años, aunque, como se mencionó antes, etiquetar cada área de forma individual puede ser complicado, especialmente en un gráfico con muchas categorías como este.

Un diagrama de caja proporciona una representación visual de la distribución de los datos y es particularmente útil para comparar distribuciones entre varios grupos. Aquí explico cómo interpretarlo:

__Elementos de un Diagrama de Caja__

1. **La Caja**:
   - **La línea inferior de la caja** representa el primer cuartil (Q1), es decir, el 25% de los datos están por debajo de este valor.
   - **La línea superior de la caja** representa el tercer cuartil (Q3), lo que significa que el 75% de los datos están por debajo de este valor.
   - **La línea dentro de la caja** es la mediana (Q2), indicando el valor medio de los datos.

2. **Los Bigotes (Whiskers)**:
   - Se extienden desde la parte superior e inferior de la caja hasta los valores máximos y mínimos dentro de un rango determinado (normalmente 1.5 veces el rango intercuartílico desde Q1 y Q3). Indican la variabilidad fuera de los cuartiles superior e inferior.

3. **Puntos Fuera de los Bigotes**:
   - Representan valores atípicos, es decir, observaciones que caen fuera del rango típico de los datos.

Dado que se están comparando medias, una prueba t de Student para muestras independientes es adecuada. El valor de umbral alfa se establecerá en 0.05, que es común en estudios estadísticos, indicando un 5% de tolerancia al error tipo I (rechazar la hipótesis nula cuando es verdadera).

   - Hipótesis alternativa (H1): Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes no son iguales.

Para probar estas hipótesis, se realiza una prueba t de Student para muestras independientes. Esta prueba es adecuada ya que compara las medias de dos grupos independientes y es robusta incluso si las varianzas no son iguales, gracias al ajuste de Welch que se suele aplicar. Ahora se realizarán las pruebas estadísticas usando los datos de las medias y desviaciones estándar.
